```python
from IPython.display import clear_output
import os
clear_output()

def neues_spiel():
    frage = input('\nWollen Sie TikTakToe spielen? j oder n.\n\n').upper()    
    if frage == 'J':        
        spieler = ''    
        while spieler != 'X' and spieler != 'O':        
            spieler = input('\nSpieler 1: Willst Du X oder O sein?\n\n').upper()
    else:
        os._exit(0)
    clear_output()
    return spieler
        
def ausgabe(reihe1, reihe2, reihe3):
    '''Ausgabe des Spielboards'''
    print(f'\nSpieler 1 ist {spieler1}')
    print(f'Spieler 2 ist {spieler2}\n')
    print('\nDas aktuelle Bord\n')
    print(' ' + reihe1[0] + ' ' + '|' + ' ' + reihe1[1] + ' ' + '|' + ' ' + reihe1[2] + ' ')
    print('-' + '-' + '-' + '|' + '-' + '-' + '-' + '|' + '-' + '-' + '-')
    print(' ' + reihe2[0] + ' ' + '|' + ' ' + reihe2[1] + ' ' + '|' + ' ' + reihe2[2] + ' ')
    print('-' + '-' + '-' + '|' + '-' + '-' + '-' + '|' + '-' + '-' + '-')
    print(' ' + reihe3[0] + ' ' + '|' + ' ' + reihe3[1] + ' ' + '|' + ' ' + reihe3[2] + ' ')
    print('\nDie Zahlen des Spielbretts')
    feldzahlen()
    
def feldzahlen():
    print('')
    print(' ' + '1' + ' ' + '|' + ' ' + '2' + ' ' + '|' + ' ' + '3' + ' ')
    print('-' + '-' + '-' + '|' + '-' + '-' + '-' + '|' + '-' + '-' + '-')
    print(' ' + '4' + ' ' + '|' + ' ' + '5' + ' ' + '|' + ' ' + '6' + ' ')
    print('-' + '-' + '-' + '|' + '-' + '-' + '-' + '|' + '-' + '-' + '-')
    print(' ' + '7' + ' ' + '|' + ' ' + '8' + ' ' + '|' + ' ' + '9' + ' ')
        
def benutzereingabe(spieler):
    # Feldeingabe des Spielers
    eingabe = 'Falsch'
    bereich = range(1,10)
    
    while eingabe.isdigit() == False or int(eingabe) not in bereich:
        eingabe = input(f'\nSpieler {spieler} bitte eine gültige noch nicht vergebene Feldnummer eingeben: ')
                
        if eingabe.isdigit() == False or int(eingabe) not in bereich:
            clear_output()
            print('\nGültige Felder sind 1 bis 9')            
    else:
        return int(eingabe)
    
def pruefen_auf_leeres_feld(zug, board):
    korrekter_zug = zug.issubset(board)
    return korrekter_zug
    
def stein_setzen (spieler, board):
    gesetzt = False
    while gesetzt == False:
        spieler_zug = benutzereingabe(spieler)
        if not pruefen_auf_leeres_feld({spieler_zug}, besetzte_felder):
            board.add(spieler_zug)
            besetzte_felder.add(spieler_zug)
            if spieler_zug <= 3:
                reihe1[spieler_zug - 1] = spieler
            elif spieler_zug <= 6:
                reihe2[spieler_zug - 4] = spieler
            else:
                reihe3[spieler_zug - 7] = spieler
            gesetzt = True
            
gewonnen_mit = ({1,2,3}, {4,5,6}, {7,8,9}, {1,4,7}, {2,5,8}, {3,6,9}, {1,5,9}, {3,5,7})

besetzte_felder_1 = set()
besetzte_felder_2 = set()
besetzte_felder = set()

reihe1 = [' ', ' ', ' ']
reihe2 = [' ', ' ', ' ']
reihe3 = [' ', ' ', ' ']

spielende = False
gewinner = ''
zaehler = 0
spieler1 = ''
spieler2 = ''
    
spieler = neues_spiel()

if spieler == 'X':
    spieler1, spieler2 = ('X', 'O')
else:
    spieler1, spieler2 = ('O', 'X')
    
clear_output()
ausgabe(reihe1, reihe2, reihe3)
    
while spielende == False and zaehler < 5:
    stein_setzen(spieler1, besetzte_felder_1)
    clear_output()
    ausgabe(reihe1, reihe2, reihe3)
    zaehler += 1
    for element in gewonnen_mit:
        if element.issubset(besetzte_felder_1):
            gewinner = spieler1
            spielende = True
    
    if zaehler == 5:
        clear_output()
        ausgabe(reihe1, reihe2, reihe3)
        print('\nSpielende. Kein Gewinner')
        break
    
    stein_setzen(spieler2, besetzte_felder_2)
    clear_output()
    ausgabe(reihe1, reihe2, reihe3)
    for element in gewonnen_mit:
        if element.issubset(besetzte_felder_2):
            if gewinner == spieler1:
                gewinner = 'unentschieden'
            else:
                gewinner = spieler2
            spielende = True
if spielende:
    if gewinner == 'unentschieden':
        clear_output()
        ausgabe(reihe1, reihe2, reihe3)
        print('\nDas Spiel endet unentschieden')
    else:
        clear_output()
        ausgabe(reihe1, reihe2, reihe3)
        print(f'\nSpieler {gewinner} hat gewonnen')
```
